```python
import numpy as np
import skfuzzy as sk
import matplotlib.pyplot as plt

x = np.arange(0,11,1)

calidad = sk.trimf(x,[9,9,10])

plt.figure()
plt.plot(x,calidad,"b",linewidth=1.5,label="Servicio")

plt.title("calidad de servicio en un restaurante")
plt.ylabel("Membresia")
plt.xlabel("Nivel de Servicio")
#plt.legend(loc="center rigth",bbox_to_achor=(1.25,0.5),ncol=1,fancybox=1,shadow=1)
```
